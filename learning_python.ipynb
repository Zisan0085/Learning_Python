```python
# Task: Create a simple logic gate
birth_year = 2000 # Change this to your birth year
current_year = 2026

age = current_year - birth_year

if age >= 18:
    status = "an Adult"
else:
    status = "a Minor"

print(f"In {current_year}, you are {age} years old.")
print(f"That means you are {status}!")
```
